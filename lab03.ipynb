## 🔹 1: Errores en Python

Los errores son situaciones inesperadas que detienen la ejecución de nuestro código. Es fundamental conocer los más comunes para poder corregirlos rápidamente.

### Errores Comunes:
* **SyntaxError:** Error en la escritura del código (falta un paréntesis, comillas, etc.).
* **NameError:** Intentar usar una variable que no ha sido definida.
* **TypeError:** Operación con tipos de datos incompatibles (ej. sumar texto y números).
* **ValueError:** Valor con tipo correcto pero contenido inadecuado.
* **ZeroDivisionError:** Intentar dividir cualquier número entre cero.

print(0 / 0)           # ZeroDivisionError

print(variable_X)      # NameError

int("Hola")            # ValueError

"2" + 2               # TypeError

## 🔹 2: Excepciones (try, except, else, finally)

El manejo de excepciones nos permite "capturar" un error para que el programa no se detenga bruscamente.

### Ejemplo 1: Estructura básica try + except

```python
try:
    numero = 0
    resultado = 100 / numero
    print(f"El resultado es: {resultado}")
except ZeroDivisionError:
    print("❌ Error: No puedes dividir entre cero.")
except ValueError:
    print("❌ Error: Debes ingresar un número válido.")
```

### Ejemplo 2: Estructura completa (else y finally)

El bloque `else` se ejecuta si no hubo errores, y `finally` se ejecuta **siempre**, pase lo que pase.

```python
try:
    print("Intentando abrir un archivo...")
    # Intentamos abrir un archivo que probablemente no existe
    archivo = open("datos_inexistentes.txt", "r")
except FileNotFoundError:
    print("⚠️ Advertencia: El archivo no fue encontrado.")
else:
    print("✅ Archivo leído con éxito.")
    archivo.close()
finally:
    print("🧹 Limpieza: Finalizando operación de sistema.")
```

## 🔹 3: Uso de raise

A veces queremos forzar la aparición de un error cuando se cumple una condición lógica que no deseamos.

```python
def verificar_edad(edad):
    if edad < 0:
        raise ValueError(f"La edad no puede ser negativa: {edad}")
    return f"Edad registrada: {edad}"

try:
    print(verificar_edad(-5))
except ValueError as e:
    print(f"Capturamos el error forzado: {e}")
```

## 🔹 4: Función normal vs Función lambda

Las funciones lambda son funciones anónimas (sin nombre) que se escriben en una sola línea. Son ideales para tareas rápidas y sencillas.

```python
# Ejemplo 1: Función normal
def cuadrado_normal(n):
    return n ** 2

# Ejemplo 2: Función lambda equivalente
cuadrado_lambda = lambda n: n ** 2

print(f"Normal: {cuadrado_normal(5)}")
print(f"Lambda: {cuadrado_lambda(5)}")
```

## 🔹 5: Iteradores e Iterables

Un **iterable** es un objeto que podemos recorrer (como una lista). Un **iterador** es el agente que hace el recorrido uno a uno.

```python
# Ejemplos de Iterables
mi_lista = [10, 20]
mi_tupla = ("A", "B")
mi_cadena = "Hi"
mi_dict = {"a": 1, "b": 2}
mi_set = {100, 200}

print("Iterando una lista:")
for x in mi_lista: print(x)

# Uso manual de iter() y next()
it = iter(mi_lista) # Convertimos el iterable en un iterador
print(f"Primer elemento: {next(it)}")
print(f"Segundo elemento: {next(it)}")
# print(next(it)) # Esto daría StopIteration
```

## 🔹 6: Generadores en Python

Los generadores son funciones especiales que devuelven valores de uno en uno usando `yield` en lugar de `return`. No guardan todos los valores en memoria, lo que los hace muy eficientes.

```python
# Ejemplo 1: Función con return (devuelve todo de golpe)
def lista_numeros(n):
    res = []
    for i in range(n):
        res.append(i)
    return res

# Ejemplo 2: Generador con yield (devuelve uno por uno)
def generador_numeros(n):
    for i in range(n):
        yield i

# -----------------------------
# Aplicación del return y yield
# -----------------------------

# Usando la función normal
numeros = lista_numeros(5)
print("Función con return (lista completa):")
print(numeros)

print("\nIterando la lista con for:")
for num in numeros:
    print(num)

# -----------------------------

# Usando el generador
gen = generador_numeros(5)

print("\nGenerador con next():")
print(f"Valor 1: {next(gen)}")
print(f"Valor 2: {next(gen)}")

print("\nGenerador con for (entrega el resto):")
for num in gen:  
    print(num)

# Nota: El for continúa desde donde el generador se quedó.
```

### 📝 Ejemplo A: Aplicación con manejo de errores

Calculadora segura que maneja entradas inválidas y división por cero.

```python
def division_segura():
    try:
        a = float(input("Dividendo: "))
        b = float(input("Divisor: "))
        return a / b
    except ZeroDivisionError:
        return "❌ No se puede dividir por cero."
    except ValueError:
        return "❌ Error: Ingresa solo números."
    finally:
        print("Cálculo finalizado.")

# print(division_segura())
```

### 📝 Ejemplo B: Aplicación con iteradores e iterables

Procesar dos listas de sensores en paralelo ignorando valores nulos.

```python
sensores_temp = [22.5, 23.0, None, 24.5]
sensores_id = ["S1", "S2", "S3", "S4"]

print("Registro de temperaturas válidas:")
for sid, temp in zip(sensores_id, sensores_temp):
    if temp is None:
        print(f"   [{sid}] Sin datos - Saltando...")
        continue
    print(f"   [{sid}] Temperatura: {temp}°C")
```

### 📝 Ejemplo C: Generador de Identificadores

Crea IDs únicos de forma infinita sin ocupar memoria extra.

```python
def generador_ids():
    contador = 1
    while True:
        yield f"USER_{contador:03d}"
        contador += 1

obtener_id = generador_ids()
print(f"Nuevo ID: {next(obtener_id)}")
print(f"Nuevo ID: {next(obtener_id)}")
print(f"Nuevo ID: {next(obtener_id)}")
```
