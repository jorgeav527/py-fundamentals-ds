## 🔹 1: PARTE 1: La función map()

La función `map()` se utiliza para aplicar una función a cada elemento de un iterable, produciendo una nueva secuencia transformada.

```python
my_pets = ['alfred', 'tabitha', 'william', 'arla']
```

```python
# --- 1. La vieja escuela (con mucho tiempo libre) ---
uppered_pets = []
for pet in my_pets:
    pet_ = pet.upper()
    uppered_pets.append(pet_)
print(f"Vieja escuela: {uppered_pets}")
```

```python
# --- 2. La forma funcional (para reutilizar lógica) ---
def upper_pets(pet_list):
    res = []
    for pet in pet_list:
        res.append(pet.upper())
    return res
print(f"Funcional: {upper_pets(my_pets)}")
```

```python
# --- 3. El estilo 'Fancy' Pythonista (List Comprehension) ---
uppered_fancy = [pet.upper() for pet in my_pets]
print(f"Pythonista (List): {uppered_fancy}")
```

```python
# --- 4. El camino de map() (Trabaja de forma inteligente) ---
def upper_pet(pet):
    return pet.upper()

uppered_map = list(map(upper_pet, my_pets))
print(f"Con map(): {uppered_map}")
```

```python
# --- 5. map() con Lambda (Aún más inteligente) ---
uppered_lambda = list(map(lambda pet: pet.upper(), my_pets))
print(f"Con map + lambda: {uppered_lambda}")
```

```python
# --- 6. Generator Comprehension ---
uppered_gen = (pet.upper() for pet in my_pets)
print(f"Desde Generador: {list(uppered_gen)}")
```

## 🔹 2: PARTE 2: La función filter()

Mientras que `map()` transforma a todos, `filter()` es como el portero exigente de una discoteca: selecciona elementos basados en una condición y solo deja pasar a los que la cumplen.

```python
scores = [66, 90, 68, 59, 76, 60, 88, 74, 81, 65]
```

```python
# --- 1. La vieja escuela (bucle for + if) ---
over_75_old = []
for score in scores:
    if score > 75:
        over_75_old.append(score)
print(f"Vieja escuela: {over_75_old}")
```

```python
# --- 2. Reutilizando lógica (función de filtrado) ---
def filter_scores(score_list):
    res = []
    for s in score_list:
        if s > 75: res.append(s)
    return res
print(f"Lógica funcional: {filter_scores(scores)}")
```

```python
# --- 3. Estilo 'Fancy' Pythonista (List Comprehension) ---
over_75_fancy = [s for s in scores if s > 75]
print(f"Pythonista (List): {over_75_fancy}")
```

```python
# --- 4. El camino de filter() ---
def is_over_75(score):
    return score > 75

over_75_filter = list(filter(is_over_75, scores))
print(f"Con filter(): {over_75_filter}")
```

```python
# --- 5. filter() con Lambda ---
over_75_lambda = list(filter(lambda s: s > 75, scores))
print(f"Con filter + lambda: {over_75_lambda}")
```

```python
# --- 6. Generator Comprehension ---
over_75_gen = (s for s in scores if s > 75)
print(f"Desde Generador: {list(over_75_gen)}")
```

En Python 3, tanto `map()` como `filter()` devuelven **Iteradores**. Esto los hace increíblemente similares a los generadores por tres razones:

1.  **Eficiencia de Memoria:** No guardan todo el resultado en RAM. Puedes procesar millones de items sin que tu computadora explote.
2.  **Evaluación Perezosa:** Solo calculan el "siguiente" item cuando lo pides explícitamente.
3.  **Un solo uso:** Una vez que recorres un objeto `map` o `filter`, queda vacío.

## 🔹 3: Función zip()

Combina elementos de múltiples listas creando pares en paralelo.

```python
usuarios = ["Alice", "Bob", "Charlie"]
puntos = [85, 92, 78]

combinado = list(zip(usuarios, puntos))
print(f"Combinados: {combinado}")
```

## 🔹 4: Generadores (yield) y Eficiencia

A diferencia de las listas, los generadores no guardan todos los datos en memoria. Generan cada valor "sobre la marcha" (lazy evaluation), lo cual es ideal para grandes volúmenes de datos.

```python
import sys

# 1. Una lista ocupa mucha memoria
lista_grande = [i for i in range(10000)]
print(f"Memoria de la lista: {sys.getsizeof(lista_grande)} bytes")

# 2. Un generador ocupa memoria constante
generador_grande = (i for i in range(10000))
print(f"Memoria del generador: {sys.getsizeof(generador_grande)} bytes")

# Función generadora
def contador_infinito(inicio):
    while True:
        yield inicio
        inicio += 1

it = contador_infinito(5)
print(f"Siguiente: {next(it)}")
print(f"Siguiente: {next(it)}")
```

## 🔹 5: Introducción a Decoradores

Un decorador envuelve a otra función para añadirle funcionalidad sin tocar su código.

```python
def decorador_saludo(func):
    def envoltura():
        print("--- Inicio de la operación ---")
        func()
        print("--- Fin de la operación ---")
    return envoltura

@decorador_saludo
def procesar_datos():
    print("Procesando información importante...")

procesar_datos()
```

### 📝 Ejemplo 1: Procesamiento de Planilla Funcional

Filtra empleados con sueldos bajos y aplícales un bono, luego genera un reporte combinado.

```python
nombres = ["Ana", "Luis", "Pedro", "Marta"]
sueldos = [1500, 2500, 1800, 3000]

# 1. Filtrar quienes ganan < 2000
bajos = list(filter(lambda x: x < 2000, sueldos))

# 2. Aplicar bono del 10% a todos usando map
con_bono = list(map(lambda s: s * 1.10 if s < 2000 else s, sueldos))

# 3. Reporte final con zip
reporte = list(zip(nombres, con_bono))
print("Planilla Actualizada:")
for nombre, total in reporte:
    print(f"   - {nombre}: S/ {total:.2f}")
```
