P6. ¿Cuál es la edad promedio de aquellos que ganan >50K y <=50K? 

```python
import pandas as pd
pd.set_option('display.max_columns', None)
nba=pd.read_csv("datos.csv")
promedio_agemas50k=nba.loc[nba['salary']=='>50K','age'].mean()
print("Promedio edad >50K es ",promedio_agemas50k)
promedio_agemenos50k=nba.loc[nba['salary']=='<=50K','age'].mean()
print("Promedio edad <=50K es ",promedio_agemenos50k)
```

P7. ¿Todos los que ganan mas de 50K al año tienen un master?

```python
high_income = nba[nba['salary'] == '>50K']

# Contar el número de personas con y sin un master
total_with_master = high_income[high_income['education'] == 'Masters'].shape[0]
total_without_master = high_income[high_income['education'] != 'Masters'].shape[0]

# Calcular el porcentaje de personas con un master
percentage_with_master = (total_with_master / high_income.shape[0]) * 100

print("Porcentaje de personas con un salario mayor a 50K al año que tienen un master:", percentage_with_master)
```

P8. Determinar el número de horas promedio por semana que se trabaja en cada país para <=50K y >50K

```python
import pandas as pd
pd.set_option('display.max_columns', None)
nba=pd.read_csv("datos.csv")
for (native_country, salary) , sub_hoursperweek in nba.groupby(['native-country','salary']):
    print(native_country,salary, sub_hoursperweek['hours-per-week'].mean())
```

P9. Agrupar y contar los registros de <=50K y >50K. Graficar el resultado de manera apropiada

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que tienes un dataframe llamado 'data' que contiene la información necesaria
# 'income' representa los ingresos anuales

# Agrupar y contar los registros según los ingresos
salary_counts = nba['salary'].value_counts()

# Graficar el resultado
plt.figure(figsize=(8, 6))
salary_counts.plot(kind='bar', color=['blue', 'green'])
plt.title('Distribución de ingresos')
plt.xlabel('Ingresos')
plt.ylabel('Cantidad')
plt.xticks([0, 1], ['<=50K', '>50K'], rotation=0)
plt.show()
```
