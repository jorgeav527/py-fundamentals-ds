## 🔹 1: INTRODUCCIÓN A NUMPY

**NumPy** es una librería diseñada para realizar cálculos numéricos de alta eficiencia. Su estructura base es el `ndarray` (arreglo multidimensional o matriz).

Es fundamental entender que **Pandas se basa internamente en NumPy** para gestionar sus datos, lo que garantiza velocidad y eficiencia en el manejo de grandes volúmenes de información.

```python
import numpy as np

# Creación de un arreglo simple (matriz)
arreglo = np.array([[1, 2, 3], [4, 5, 6]])

print("Arreglo NumPy:")
print(arreglo)

# Mostrar la forma del arreglo (filas, columnas)
print(f"\nForma (shape): {arreglo.shape}")
```

## 🔹 2: PANDAS: CARGA Y MANIPULACIÓN DE DATOS

Pandas es la herramienta principal para trabajar con datos tabulares. Utilizaremos el archivo `hrdata.csv` como nuestra fuente principal.

### 📝 2.1 Carga de datos y transformación

Importamos la librería y cargamos el archivo CSV estableciendo la columna `Name` como el identificador principal (índice). Además, convertiremos las fechas a un formato que Python pueda manipular.

```python
import pandas as pd

# Lectura del archivo estableciendo el índice
df = pd.read_csv("hrdata.csv", index_col="Name")

# Conversión de la columna 'Hire Date' a tipo datetime
df["Hire Date"] = pd.to_datetime(df["Hire Date"], format="%m/%d/%y")

print("Datos cargados y transformados:")
df.head()
```

### 📝 2.2 Extracción de Componentes de Fecha

Utilizando el accesorio `.dt`, podemos descomponer la fecha en sus partes elementales para enriquecer nuestro DataFrame.

```python
# Creación de nuevas columnas a partir de la fecha de contratación
df["Year"] = df["Hire Date"].dt.year
df["Month"] = df["Hire Date"].dt.month
df["Day"] = df["Hire Date"].dt.day
df["DayName"] = df["Hire Date"].dt.day_name()

print("DataFrame actualizado con columnas de fecha:")
df.head()
```

### 📝 2.3 Inspección Básica del DataFrame

Es esencial inspeccionar los datos para entender su estructura y contenido.

```python
# 1. Dimensiones (Filas, Columnas)
print(f"Forma: {df.shape}")

# 2. Resumen técnico (tipos de datos, nulos)
print("\nInformación detallada:")
df.info()

# 3. Primeras 5 filas por defecto
print("\nPrimeras filas:")
print(df.head())

# 4. Muestra aleatoria de 2 registros
print("\nMuestra aleatoria:")
print(df.sample(2))

# 5. Estadísticas descriptivas básicas
print("\nResumen estadístico:")
print(df.describe())
```

## 🔹 3: MANEJO DE SERIES

Una **Series** es un arreglo unidimensional capaz de contener cualquier tipo de datos. Esencialmente, es una columna individual.

```python
# Creación de una Series con índices personalizados
obj2 = pd.Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])

print("Series obj2:")
print(obj2)

# Acceso a elementos por etiqueta
print(f"\nValor en el índice 'd': {obj2['d']}")

# Aplicación de máscaras booleanas y comparaciones
print("\nValores en obj2 que son mayores a 0:")
print(obj2[obj2 > 0])
```

## 🔹 4: DATAFRAMES: CREACIÓN Y OPERACIONES

El **DataFrame** representa una estructura de datos bidimensional (tablas) con colecciones de columnas ordenadas.

### 📝 4.1 Creación del DataFrame

Podemos construir un DataFrame desde un diccionario de Python.

```python
data = {
    "state": ["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
    "year": [2000, 2001, 2002, 2001, 2002, 2003],
    "pop": [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]
}
df_states = pd.DataFrame(data)

print("DataFrame creado:")
print(df_states)
```

### 📝 4.2 Reordenamiento de Columnas

Podemos reorganizar las columnas pasando una lista con el orden deseado.

```python
# Definir el nuevo orden de columnas
cols = ["year", "state", "pop"]
print("DataFrame reordenado:")
print(df_states[cols])
```

### 📝 4.3 Acceso a Columnas

Existen dos formas principales de acceder a los datos de una columna específica.

```python
# Acceso con corchetes (clave de diccionario)
print("Columna 'state' (Forma corchetes):")
print(df_states["state"])

# Acceso con punto (atributo)
print("\nColumna 'state' (Forma punto):")
print(df_states.state)
```

### 📝 4.4 Agregar una Columna

Es posible insertar nuevas columnas con valores constantes o secuencias.

```python
# Agregar columna 'dept' con valores incrementales
df_states["dept"] = np.arange(6.)

print("DataFrame con nueva columna:")
print(df_states)
```

### 📝 4.5 Eliminar Filas o Columnas

El método `.drop()` se utiliza para remover elementos. Por defecto devuelve una **copia** del DataFrame; si queremos modificar el original debemos usar `inplace=True`.

*   `axis=0`: Indica que se eliminarán filas.
*   `axis=1`: Indica que se eliminarán columnas.

```python
# DataFrame de prueba para eliminación
demo_df = pd.DataFrame(np.arange(16).reshape((4, 4)),
                      index=["Ohio", "Colorado", "Utah", "New York"],
                      columns=["one", "two", "three", "four"])

# 1. Eliminar fila 'Colorado' (devuelve una copia)
print("Resultado de eliminar 'Colorado' (copia):")
print(demo_df.drop("Colorado"))

# 2. Eliminar fila 'Ohio' modificando el original (inplace)
demo_df.drop("Ohio", inplace=True)

# 3. Eliminar columnas usando axis=1
limpio_df = demo_df.drop(["two", "four"], axis=1)

print("\nDataFrame final después de las eliminaciones:")
print(limpio_df)
```

### 📝 4.6 Selección con loc e iloc

`.loc` realiza selecciones basadas en **etiquetas**, mientras que `.iloc` se basa en **posiciones enteras**.

```python
# Selección por etiqueta con .loc
print("Selección de la fila 'Utah' con .loc:")
print(limpio_df.loc["Utah"])

# Selección por posición con .iloc
print("\nSelección de la primera fila con .iloc:")
print(limpio_df.iloc[0])
```

## 🔹 5: MÁSCARAS Y FILTROS

Los filtros nos permiten extraer subconjuntos de datos que cumplen con criterios lógicos específicos.

```python
# 1. Máscara booleana simple
print("Máscara (Salario > 60k):")
mask = df["Salary"] > 60000
print(df[mask])

# 2. Filtros con condiciones compuestas (& = y)
filtro_comp = df[(df["Salary"] > 50000) & (df["Sick Days remaining"] > 5)]
print("\nResultado de filtro compuesto:")
print(filtro_comp)

# 3. Selección específica con .loc y máscara
print("\nSueldos de empleados contratados después del 2013:")
print(df.loc[df["Year"] > 2013, ["Salary"]])

# 4. Uso pequeño de lambda para filtrar
print("\nFiltro rápido con lambda (Sueldo > 65k):")
print(df[lambda x: x["Salary"] > 65000])
```

## 🔹 6: ORDENAMIENTO (SORT)

Podemos organizar las filas de una tabla basándonos en sus valores o en sus índices.

```python
# Ordenar por valores de una columna (ej. Salario descendente)
print("Ordenado por Salario (Descendente):")
print(df.sort_values(by="Salary", ascending=False))

# Ordenar por el índice (Alfabético por nombre)
print("\nOrdenado por Índice:")
print(df.sort_index())
```
