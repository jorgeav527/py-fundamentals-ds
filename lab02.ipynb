## 🔹 Control de Flujo

### 🔹 1: Estructuras Condicionales (if, elif, else)

Permiten ejecutar bloques de código basados en condiciones. En una estructura `if-elif-else`, **solo un bloque** se ejecutará (el primero cuya condición sea verdadera).

```python
some_var = 5

if some_var > 10:
    print("some_var es totalmente mayor que 10.")
elif some_var < 10: 
    print("some_var es menor que 10.") # Este bloque se ejecuta
else: 
    print("some_var es efectivamente 10.")
```

### 🔹 2: Comparación de Estructuras (if independientes)

A diferencia del caso anterior, si usamos múltiples `if` seguidos, Python evaluará **cada uno** de ellos de forma independiente. Esto es útil cuando queremos que ocurran varias acciones si se cumplen varias condiciones.

```python
puntaje = 15

print("--- Evaluación con if-elif ---")
if puntaje >= 10:
    print("Aprobado")
elif puntaje >= 15:
    print("Excelente") # Nunca llegará aquí porque el primero ya fue True

print("\n--- Evaluación con if-if independientes ---")
if puntaje >= 10:
    print("Aprobado")
if puntaje >= 15:
    print("Excelente") # Ambos se ejecutan porque son independientes
```

### 🔹 3: Bucle for

Se utiliza para iterar sobre una secuencia (lista, tupla, diccionario, set o string).

```python
for animal in ["dog", "cat", "mouse"]:
    print(f"{animal} es un mamífero")

# Uso de range para secuencias numéricas
for i in range(4):
    print(f"Iteración: {i}")
```

### 🔹 4: Bucle while con Control (break y continue)

El bucle `while` se ejecuta mientras una condición sea verdadera. 
- `continue`: Salta el resto del código en la iteración actual y vuelve al inicio del bucle.
- `break`: Rompe el bucle por completo.

```python
x = 0
print("Bucle while con control de flujo:")
while x < 10:
    x += 1
    
    if x == 3:
        print("   Saltamos el 3 con 'continue'")
        continue
    
    if x == 7:
        print("   Rompemos el bucle en 7 con 'break'")
        break
    
    print(f"   Valor actual: {x}")
```

## 🔹 Funciones

### 🔹 5: Funciones Básicas (Sin y Con Tipeo)

Las funciones nos permiten encapsular lógica. Aunque Python es dinámico, el **Type Hinting** ayuda a documentar qué esperamos recibir y qué retornamos.

```python
# Función básica sin tipeo
def sumar_basico(x, y):
    return x + y

# Función con Type Hinting (Tipeo)
def sumar_tipado(x: int, y: int) -> int:
    """Suma dos números enteros y retorna un entero."""
    return x + y

print(f"Resultado sin tipeo: {sumar_basico(5, 6)}")
print(f"Resultado con tipeo: {sumar_tipado(10, 20)}")
```

### 🔹 6: Argumentos Flexibles (*args y **kwargs)

- `*args`: Recibe una cantidad arbitraria de argumentos posicionales (como una tupla).
- `**kwargs`: Recibe una cantidad arbitraria de argumentos de palabra clave (como un diccionario).

```python
def funcion_flexible(*args, **kwargs) -> None:
    print(f"Argumentos posicionales (*args): {args}")
    print(f"Argumentos de clave (**kwargs): {kwargs}")

funcion_flexible(1, 2, "hola", ciudad="Lima", activo=True)
```

### 🔹 7: El Scope (Alcance) de las Variables

Es crucial entender dónde vive una variable. Una variable definida dentro de una función es **local** y no afecta a las variables **globales** a menos que se indique explícitamente.

```python
x = 5 # Variable Global

def intentar_cambiar_x(nuevo_valor):
    x = nuevo_valor  # Crea una nueva variable LOCAL llamada x
    print(f"[Dentro de set_x] x local es: {x}")
    
def cambiar_x_global(nuevo_valor):
    global x # Indica que usaremos la variable global
    print(f"[Dentro de set_global_x] Antes de cambiar: {x}")
    x = nuevo_valor 
    print(f"[Dentro de set_global_x] Después de cambiar: {x}")
    
print("Valor de x antes de las funciones:", x)
intentar_cambiar_x(43)
print("Valor de x después de intentar_cambiar_x:", x, "(No cambió el global)")

cambiar_x_global(6)
print("Valor de x después de set_global_x:", x, "(Sí cambió el global)")
```

### 📝 Problema 1: El Juego del FizzBuzz

Implementa la lógica que imprime números del 1 al 15. 
- Si es múltiplo de 3, imprime "FIZZ".
- Si es múltiplo de 5, imprime "BUZZ".
- Si es múltiplo de ambos, imprime "FIZZBUZZ".

```python
for i in range(1, 16):
    if i % 3 == 0 and i % 5 == 0:
        print('FIZZBUZZ')
    elif i % 3 == 0:
        print('FIZZ')
    elif i % 5 == 0:
        print('BUZZ')
    else:
        print(i)
```

### 📝 Problema 2: Buscador con Filtros y Límites

Imagina que tienes una lista de códigos de productos. Queremos buscar uno específico, pero debemos **saltar** los códigos defectuosos y **detener** la búsqueda si encontramos un código de seguridad crítico.

**Objetivo:** Crear una función que use una variable global para contar cuántas veces se ha ejecutado el buscador, use `continue` para ignorar errores y `break` para seguridad.

```python
total_ejecuciones = 0  # Variable en Scope Global

def buscador_inteligente(lista, objetivo, error, critico):
    global total_ejecuciones
    total_ejecuciones += 1
    
    print(f"--- Ejecución nro {total_ejecuciones} ---")
    encontrado = False
    
    for codigo in lista:
        if codigo == error:
            print(f"   [SKIP] Código {codigo} defectuoso detectado. Saltando...")
            continue  # Salta a la siguiente iteración
            
        if codigo == critico:
            print(f"   [STOP] Código crítico {codigo} encontrado! Deteniendo búsqueda por seguridad.")
            break  # Sale del bucle inmediatamente
            
        print(f"   Analizando: {codigo}")
        if codigo == objetivo:
            encontrado = True
            break
            
    return encontrado

# Datos de prueba
codigos_almacen = [101, 404, 105, 999, 202]
target = 105
error_code = 404
security_halt = 999

res = buscador_inteligente(codigos_almacen, target, error_code, security_halt)
print(f"\n¿Producto encontrado?: {res}")
print(f"Total de búsquedas globales: {total_ejecuciones}")
```
