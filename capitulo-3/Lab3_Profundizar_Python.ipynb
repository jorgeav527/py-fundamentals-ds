## 1. Abrir archivos (open)

````python
# Abrimos el archivo "dog_breeds.txt" en modo lectura ("r").
file = open("dog_breeds.txt", "r")

# Leemos el contenido del archivo y lo almacenamos en la variable 'content'.
content = file.read()

# Cerramos el archivo para liberar recursos.
file.close()

# Mostramos el contenido del archivo, independientemente de si ocurren excepciones o no.
print(content)
````

````python
try:
    # Intentamos abrir el archivo "dog_breeds.txt" de manera segura.
    with open("dog_breeds.txt", "r") as file:
        # Leemos el contenido del archivo y lo almacenamos en la variable 'content'.
        content = file.read()
except FileNotFoundError:
    # Si el archivo no se encuentra, mostramos un mensaje de error.
    print("El archivo 'dog_breeds.txt' no se encuentra en el directorio actual.")
except Exception as e:
    # Para otros errores no controlados, mostramos un mensaje de error general.
    print(f"Se ha producido un error: {e}")
finally:
    # En el bloque 'finally', mostramos el contenido del archivo,
    # independientemente de si se produjo una excepción o no.
    print("Contenido del archivo (independientemente de las excepciones):\n")
    print(content)
````

````python
# Abrir el archivo "dog_breeds.txt" en modo binario de lectura ("rb") con búfer de tamaño 0
file = open("dog_breeds.txt", "rb", buffering=0)

# Leer el contenido binario del archivo y almacenarlo en la variable 'binary_content'
binary_content = file.read()

# Cerrar el archivo
file.close()

# Imprimir el contenido binario
print(binary_content)
````

El "búfer" se refiere a una técnica utilizada para mejorar el rendimiento de la lectura o escritura de datos en archivos. En la programación, cuando leemos o escribimos datos en un archivo, estos datos a menudo se almacenan temporalmente en un búfer antes de transferirse al disco o desde el disco al programa.

En términos sencillos, con buffering=0, cada vez que se lee desde el archivo, los datos se leen directamente del archivo y se entregan al programa sin ningún tipo de almacenamiento temporal. Esto es útil en situaciones donde necesitas acceder a los datos del archivo de inmediato y no deseas esperar a que se acumule una cantidad significativa de datos en el búfer.

````python
# Abrir el archivo "dog_breeds.txt" en modo lectura ("r") con 'with'
with open("dog_breeds.txt", "r") as file:
    # Inicializar una variable para almacenar cada línea leída
    line = file.readline()

    # Usar un bucle 'while' para iterar hasta que no haya más líneas para leer
    while line:
        # Imprimir la línea actual
        print(line, end='')
        
        # Leer la siguiente línea
        line = file.readline()
````

## 6. Escritura.

````python
# Abrir el archivo "dog_breeds.txt" en modo lectura ("r") con 'with'
with open("dog_breeds.txt", "r") as original_file:
    # Leer todas las líneas del archivo original en una lista
    lines = original_file.readlines()

# Abrir un nuevo archivo "reversed_dog_breeds.txt" en modo escritura ("w") con 'with'
with open("reversed_dog_breeds.txt", "w") as reversed_file:
    # Iterar a través de las líneas en orden inverso y escribirlas en el nuevo archivo
    for line in reversed(lines):
        reversed_file.write(line)
````

- vamos a realizar un análisis básico de los datos JSON para determinar qué usuarios han completado la mayor cantidad de tareas y muestra una lista de usuarios ordenados por las tareas completadas.

````python
# Mapeos con TODOs
 
todos_by_user = {}  # Un diccionario para almacenar el recuento de tareas completadas por usuario

# Iteramos a través de cada tarea en la lista de todos los datos
for todo in todos:
    if todo["completed"]:
        # Verificamos si la tarea está marcada como completada (campo "completed" en True)
        try:
            # Intentamos incrementar el recuento de tareas completadas para el usuario actual
            todos_by_user[todo["userId"]] += 1
        except KeyError:
            # Si el usuario no existe en el diccionario, lo agregamos con un recuento de 1
            todos_by_user[todo["userId"]] = 1

# Ordenamos a los usuarios por la cantidad de tareas completadas en orden descendente
top_users = sorted(todos_by_user.items(), key=lambda x: x[1], reverse=True)

# Imprimimos la lista de los usuarios con la cantidad de tareas completadas
print(top_users)
````

- Ahora vamos a encuentrar a los usuarios con la mayor cantidad de tareas completadas y los imprime en un formato legible. 

````python
# Caso TODOs para Operaciones.

# Obtener la cantidad de tareas completadas por el usuario con la mayor cantidad de tareas completadas
max_complete = top_users[0][1]

# Crear una lista para almacenar los IDs de los usuarios con la mayor cantidad de tareas completadas
users = []

# Iterar a través de la lista 'top_users'
for user, num_complete in top_users:
    # Verificar si el número de tareas completadas es igual al máximo
    if num_complete < max_complete:
        # Si no es igual, salir del bucle (los usuarios están ordenados por cantidad, los siguientes tendrán menos)
        break
    # Agregar el ID del usuario a la lista 'users'
    users.append(str(user))

# Crear una cadena que combine los IDs de los usuarios con la mayor cantidad de tareas completadas, separados por "and"
max_users = " and ".join(users)

# Imprimir la lista de usuarios con la mayor cantidad de tareas completadas
print(max_users)
````

````python
# Paso 1: Importa las bibliotecas necesarias
from sqlalchemy import create_engine, Column, Integer, String, ForeignKey
from sqlalchemy.orm import sessionmaker, relationship, declarative_base

# Paso 2: Crea el motor de SQLite y la sesión de SQLAlchemy
engine = create_engine('sqlite:///mydatabase.db', echo=False)  # 'mydatabase.db' es el nombre de tu archivo de base de datos
Session = sessionmaker(bind=engine)
session = Session()

# Paso 3: Define una clase base declarativa
Base = declarative_base()

# Paso 4: Define las clases que representarán tus tablas
class User(Base):
    __tablename__ = 'users'
    id = Column(Integer, primary_key=True)
    username = Column(String)

    # Relación con la tabla 'posts'
    posts = relationship('Post', back_populates='user')

class Post(Base):
    __tablename__ = 'posts'
    id = Column(Integer, primary_key=True)
    title = Column(String)
    content = Column(String)
    user_id = Column(Integer, ForeignKey('users.id'))

    # Relación con la tabla 'users'
    user = relationship('User', back_populates='posts')

    # Relación con la tabla 'comments'
    comments = relationship('Comment', back_populates='post')

class Comment(Base):
    __tablename__ = 'comments'
    id = Column(Integer, primary_key=True)
    text = Column(String)
    post_id = Column(Integer, ForeignKey('posts.id'))

    # Relación con la tabla 'posts'
    post = relationship('Post', back_populates='comments')

# Paso 5: Crea las tablas en la base de datos
Base.metadata.create_all(engine)

# Paso 6: Define funciones para crear, actualizar, borrar y leer datos
def create_user(username):
    new_user = User(username=username)
    session.add(new_user)
    session.commit()

def create_post(title, content, user_id):
    new_post = Post(title=title, content=content, user_id=user_id)
    session.add(new_post)
    session.commit()

def create_comment(text, post_id):
    new_comment = Comment(text=text, post_id=post_id)
    session.add(new_comment)
    session.commit()

def read_users():
    users = session.query(User).all()
    return users

def read_posts():
    posts = session.query(Post).all()
    return posts

def read_comments():
    comments = session.query(Comment).all()
    return comments

def update_user(user_id, new_username):
    user = session.query(User).filter_by(id=user_id).first()
    user.username = new_username
    session.commit()

def delete_user(user_id):
    user = session.query(User).filter_by(id=user_id).first()
    session.delete(user)
    session.commit()
````

````python
# Paso 7: Ejecuta las funciones
create_user('Alice')
create_user('Bob')
create_post('First Post', 'Hello, World.', 1)
create_post('Second Post', 'This is a test.', 2)
create_post('Third Post', 'Testing test.', 2)
create_comment('Great post!', 1)
create_comment('Bad comment!', 1)
create_comment('Nice job!', 2)
create_comment('New comment!', 2)
create_comment('Bad comment!', 2)
````

### Query CLI

`SELECT p.title, COUNT(c.id) AS num_comments FROM posts p LEFT JOIN comments c ON p.id = c.post_id GROUP BY p.id, p.title ORDER BY num_comments DESC;`

- SELECT p.title, COUNT(c.id) AS num_comments: Selecciona el título de cada publicación (post) de la tabla "posts" y cuenta la cantidad de comentarios (comments) de la tabla "comments". El resultado se almacena con el nombre "num_comments" para mayor claridad.

- FROM posts p: Indica que los datos provienen de la tabla "posts" y se le da un alias "p" para facilitar su uso en la consulta.

- LEFT JOIN comments c ON p.id = c.post_id: Realiza una unión izquierda (LEFT JOIN) entre las tablas "posts" y "comments". Esto significa que se incluirán todas las filas de "posts" y las filas correspondientes de "comments" si coinciden en la condición de que el "id" de "posts" sea igual al "post_id" de "comments".

- GROUP BY p.title: Agrupa los resultados por el campo "title" de "posts". Esto significa que se contarán los comentarios para cada título de publicación en lugar de mostrar cada comentario individualmente.

- ORDER BY num_comments DESC: Ordena los resultados en orden descendente según el número de comentarios. De esta manera, las publicaciones con más comentarios aparecerán primero en los resultados.

```md
title            num_comments
---------------  ------------
Second Post      3           
First Post       2           
Título del Post  1           
Third Post       0 
```

### Query Python and SQLalchemy

- Selecciona el título de cada publicación (columna title) de la tabla posts.
- Cuenta la cantidad de comentarios para cada publicación utilizando una unión izquierda (LEFT JOIN) con la tabla comments y agrupa los resultados por el ID de la publicación (p.id) y su título (p.title).
- Ordena el resultado en orden descendente según la cantidad de comentarios (num_comments) para que las publicaciones con más comentarios aparezcan primero.
- Puedes usar esta consulta como punto de partida para discutir temas como JOINs, GROUP BY y ORDER BY con tus alumnos.

````python
from sqlalchemy import func
from sqlalchemy.orm import aliased

def get_post_comments_count():
    post_alias = aliased(Post)
    
    query = (
        session.query(
            post_alias.title,
            func.count(Comment.id).label('num_comments')
        )
        .outerjoin(Comment, post_alias.id == Comment.post_id)
        .group_by(post_alias.id, post_alias.title)
        .order_by(func.count(Comment.id).desc())
    )

    return query.all()
````

````python
post_comments_count = get_post_comments_count()

for post, num_comments in post_comments_count:
    print(f"{post:<15} {num_comments}")
````

### Query with a string and python

````python
from sqlalchemy import text

# Define the SQL query as a string
sql_query = """
    SELECT p.title, COUNT(c.id) AS num_comments
    FROM posts p
    LEFT JOIN comments c ON p.id = c.post_id
    GROUP BY p.title
    ORDER BY num_comments DESC;
"""

# Execute the SQL query using SQLAlchemy
result = session.execute(text(sql_query))
for row in result:
    print(row)
````

4. Diferencias entre SQL y NoSQL

- Modelo de Datos:

    - SQL: Las bases de datos SQL son bases de datos relacionales que utilizan un modelo de datos tabular o de tipo tabla. Los datos se almacenan en tablas con filas y columnas. Cada tabla tiene un esquema predefinido que define la estructura de los datos.
    - NoSQL: Las bases de datos NoSQL utilizan una variedad de modelos de datos, como documentos, columnas, grafos o pares clave-valor. No tienen un esquema fijo y permiten una flexibilidad significativa en la estructura de los datos.
 
- Escalabilidad:

    - SQL: Las bases de datos SQL suelen escalar verticalmente, lo que significa que aumentan la capacidad de una sola máquina para manejar más carga. Esto puede ser costoso y limitado en escalabilidad.
    - NoSQL: Las bases de datos NoSQL suelen escalar horizontalmente, lo que permite agregar más servidores a la infraestructura para manejar cargas más grandes. Esto proporciona una mejor escalabilidad y es más económico

- Consultas:

    - SQL: Utilizan el lenguaje SQL para consultas estructuradas y complejas. Ofrecen una potente capacidad de consulta y soportan JOINs.
    - NoSQL: Utilizan modelos de consulta específicos para el tipo de base de datos NoSQL (por ejemplo, consultas basadas en documentos o consultas de grafos). Las capacidades de consulta pueden variar según el tipo de NoSQL.

- Transacciones:

    - SQL: Las bases de datos SQL son ACID (Atomicidad, Consistencia, Aislamiento, Durabilidad) y son ideales para aplicaciones que requieren transacciones robustas.
    - NoSQL: Las bases de datos NoSQL a menudo sacrifican algunas propiedades ACID en favor de un rendimiento y escalabilidad mejorados. Algunas bases de datos NoSQL admiten ACID, pero esto puede variar.

- Uso Adecuado:

    - SQL: Adecuado para aplicaciones que requieren una estructura de datos rígida y transacciones complejas, como aplicaciones empresariales tradicionales.
    - NoSQL: Adecuado para aplicaciones con datos no estructurados o semiestructurados, escalabilidad masiva, alta velocidad y flexibilidad en la estructura de datos, como aplicaciones web, redes sociales, IoT y análisis de datos.

Ejemplos:

SQL: MySQL, PostgreSQL, Oracle, SQL Server.
NoSQL: MongoDB, Cassandra, Redis, Neo4j.
