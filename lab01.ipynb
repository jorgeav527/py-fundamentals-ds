## 🔹 Introducción a las Variables

Una variable en Python es un nombre que apunta a un valor almacenado en la memoria. Podemos guardar diferentes tipos de datos, como texto, números o estados lógicos.

### Reglas para nombres de variables en Python

### ✔ Permitido:
* **Letras:** (a-z, A-Z)
* **Números:** (0-9), pero **NO** al inicio.
* **Guion bajo:** (_)

### ✔ Convención para constantes globales:
Se recomienda seguir este estilo para valores que no cambian:
* **TODO EN MAYÚSCULAS**
* Palabras separadas con **GUIONES BAJOS**.
    * *Ejemplos:* MAX_USUARIOS, API_KEY, CONFIG_PATH

### ✘ NO permitido:
* **Empezar con números:** (e.g., 1variable)
* **Usar espacios:** (e.g., nombre completo)
* **Usar símbolos especiales:** (e.g., nombre$, dato#, valor-1)
* **Usar palabras reservadas:** (e.g., for, class, if, return)

---

### Ejemplos válidos:
- **Variables estándar:** nombre, edad, activo, igv, _contador, dato2, nombre_completo
- **Constantes globales:** MAX_VALOR, CONFIG_PATH, API_VERSION

```python
# No válidos

import keyword
import builtins

palabras_reservadas = keyword.kwlist
nombres_builtin = dir(builtins)

print(palabras_reservadas)
print(nombres_builtin)
```

```python
# Ejemplos básicos de variables
nombre = "jorge_alarcon"
edad = 25
activo = True
igv = 0.18

print(f"Nombre: {nombre}")
print(f"Edad: {edad}")
print(f"Activo: {activo}")
print(f"igv: {igv}")
```

```python
nombre = "jorge_alarcon"

# Inspeccionamos los métodos disponibles para un objeto tipo String
metodos_string = set(dir(nombre))
print(metodos_string)
```

## 🔹 Sección: Primitivos

### 🔹 1: Operaciones con números (enteros)

Los números enteros (tipo `int`) se utilizan para contar o representar cantidades exactas.

```python
numero = 10
```

Aunque no solemos usar métodos con números frecuentemente, Python nos permite realizar operaciones matemáticas básicas de forma muy sencilla:

```python
print(numero + 5)      # Suma
print(numero * 2)      # Multiplicación
print(numero - 3)      # Resta

print(numero / 3)      # División normal (float)
print(numero // 3)     # División entera (sin decimales)
print(numero % 3)      # Módulo (residuo de la división)

print(numero ** 2)     # Exponente (potencia)
print(numero ** (1/2))   # Raíz cuadrada usando potencia
```

---

### 🔹 2: Números flotantes

Los números flotantes (tipo `float`) representan valores con decimales.

```python
numero = 10.5

print(numero + 5)        # Suma
print(numero * 2)        # Multiplicación
print(numero - 3)        # Resta

print(numero / 3)        # División normal (float)
print(numero // 3)       # División entera (trunca decimales)
print(numero % 3)        # Módulo (residuo de la división)

print(numero ** 2)       # Exponente (potencia)
print(numero ** (1/2))   # Raíz cuadrada usando potencia
```

### 🔹 3: Booleanos

Los booleanos (tipo `bool`) representan valores de verdad: `True` (Verdadero) o `False` (Falso).

```python
es_activo = True
```

```python
# Comparaciones básicas
print(1 < 10)   # True: 1 es menor que 10
print(1 > 10)   # False: 1 no es mayor que 10
print(2 <= 2)   # True: 2 es menor o igual que 2
print(2 >= 2)   # True: 2 es mayor o igual que 2

# Comparaciones usando operadores lógicos
print(1 < 2 and 2 < 3)  # True: ambas comparaciones son verdaderas
print(2 < 3 and 3 < 2)  # False: la segunda comparación es falsa

# Comparaciones encadenadas (Python permite esto)
print(1 < 2 < 3)  # True: 1 < 2 y 2 < 3
print(2 < 3 < 2)  # False: 3 no es menor que 2
```

```python
es_activo = True
edad = 18

# --- Comparaciones ---
print(edad > 10)     # Mayor que
print(edad == 18)    # Igualdad

# --- Operaciones lógicas ---
print(es_activo and False)   # Operación AND (y)
print(es_activo or False)    # Operación OR (o)
print(not es_activo)         # Negación lógica (NOT)
```

### 🔹 4: Strings

Los strings (tipo `str`) son cadenas de texto. Son objetos muy versátiles con muchos métodos útiles para manipular texto.

```python
texto = "hola_mundo"
```

```python
# Métodos de strings más utilizados
print(texto.upper())             # Convierte todo a MAYÚSCULAS
print(texto.lower())             # Convierte todo a minúsculas
print(texto.capitalize())        # Primera letra en mayúscula
print(texto.title())             # Primera letra de cada palabra en mayúscula
print(texto.swapcase())          # Invierte mayúsculas/minúsculas

print(texto.replace("_", " "))   # Reemplaza un carácter por otro
print(texto.startswith("hola"))  # Verifica si empieza con 'hola'
print(texto.endswith("mundo"))   # Verifica si termina con 'mundo'

print(texto.find("mun"))         # Devuelve posición de 'mun' (o -1 si no existe)
print(texto.index("mun"))        # Igual que find, pero da error si no existe

print(texto.split("_"))          # Divide el texto en una lista usando "_"
print(" ".join(["hola", "mundo"]))  # Une elementos de una lista con un espacio

print(texto.isalpha())           # ¿Solo letras? (falso por el '_')
print(texto.isdigit())           # ¿Solo números?
print(texto.islower())           # ¿Está todo en minúsculas?
print(texto.isupper())           # ¿Está todo en MAYÚSCULAS?
print(texto.isidentifier())      # ¿Cumple reglas de nombre de variable?
print(texto.isascii())           # ¿Contiene solo caracteres ASCII?

print(texto.rjust(15))           # Alinea a la derecha en 15 espacios
print(texto.ljust(15))           # Alinea a la izquierda en 15 espacios
print(texto.center(20, "-"))     # Centra el texto usando '-' como relleno

print(texto.strip("ho"))         # Elimina caracteres del inicio y final
print(texto.removeprefix("hola_"))  # Elimina prefijo (Python 3.9+)
print(texto.removesuffix("mundo"))  # Elimina sufijo (Python 3.9+)
```

## 🔹 Sección: Colecciones

Las colecciones nos permiten agrupar múltiples valores en una sola variable.

### 🔹 5: Listas

Las listas son colecciones ordenadas y **mutables** (pueden cambiar).

```python
lista = [1, 2, 3]

# Agregar elementos
lista.append(4)              # Agrega un elemento al final
print(lista)

lista.extend([5, 6])         # Agrega varios elementos
print(lista)

lista.insert(0, 0)           # Inserta un elemento en una posición específica
print(lista)

# Eliminar elementos
lista.remove(3)              # Elimina el valor 3
print(lista)

ultimo = lista.pop()         # Elimina el último elemento y lo retorna
print(ultimo)
print(lista)

# Buscar información
print(lista.index(2))        # Devuelve el índice del valor 2
print(lista.count(2))        # Cuenta cuántas veces aparece el 2

# Ordenar y revertir
lista.sort()                 # Ordena la lista (ascendente)
print(lista)

lista.reverse()              # Invierte el orden
print(lista)

# Copiar listas
copia = lista.copy()         # Copia la lista
print(copia)

# Limpiar lista
lista.clear()                # Elimina todos los elementos
print(lista)
```

Acceder a los ítems de una lista en Python

```python
lista = [10, 20, 30, 40, 50]

# Acceder por índice (empieza en 0)
print(lista[0])   # Primer elemento → 10
print(lista[2])   # Tercer elemento → 30

# Índices negativos
print(lista[-1])  # Último elemento → 50
print(lista[-2])  # Penúltimo → 40

# Rebanado (slicing)
print(lista[1:4])   # Desde índice 1 hasta 3 → [20, 30, 40]
print(lista[:3])    # Desde el inicio hasta índice 2 → [10, 20, 30]
print(lista[2:])    # Desde el índice 2 hasta el final → [30, 40, 50]
print(lista[:])     # Copia completa de la lista
print(lista[::2])   # Saltos de 2 → [10, 30, 50]
```

### 🔹 6: Tuplas

Las tuplas son colecciones ordenadas pero **inmutables** (NO pueden cambiar después de su creación).

```python
tupla = (1, 2, 3, 2, 5, 2, 7)

# Las tuplas son inmutables, pero sí tienen dos métodos importantes:

print(tupla.count(2))   # Cuenta cuántas veces aparece el valor 2
print(tupla.index(5))   # Devuelve el índice donde aparece el valor 5
```

Acceder a los ítems de una tupla

```python
tupla = (10, 20, 30, 40, 50)

# Acceder por índice (igual que listas)
print(tupla[0])   # Primer elemento → 10
print(tupla[2])   # Tercer elemento → 30

# Índices negativos
print(tupla[-1])  # Último elemento → 50
print(tupla[-2])  # Penúltimo → 40

# Rebanado (slicing)
print(tupla[1:4])   # → (20, 30, 40)
print(tupla[:3])    # → (10, 20, 30)
print(tupla[2:])    # → (30, 40, 50)
print(tupla[:])     # Copia completa
print(tupla[::2])   # Saltos → (10, 30, 50)
```

### 🔹 7: Diccionarios

Los diccionarios almacenan pares de **clave: valor**.

```python
# Diccionario
diccionario = {
    "nombre": "Jorge",
    "edad": 25,
    "activo": True,
    "pais": "Perú",
    "skills": ["Python", "Go", "SQL"]
}

# Obtener valores
print(diccionario.keys())        # Obtiene todas las claves
print(diccionario.values())      # Obtiene todos los valores
print(diccionario.items())       # Obtiene clave-valor como tuplas

# Acceder de forma segura
print(diccionario.get("edad"))          # Obtiene valor sin error
print(diccionario.get("altura", "No existe"))  # Valor por defecto

# Agregar / actualizar valores
diccionario.update({"ciudad": "Lima"})  # Agrega o modifica
print(diccionario)

# Eliminar elementos
valor_eliminado = diccionario.pop("pais")  # Elimina clave y retorna su valor
print(valor_eliminado)
print(diccionario)
```

Acceder a los ítems de un diccionario

```python
print(diccionario["nombre"])   # → Jorge
print(diccionario["edad"])     # → 25
```

### 🔹 8: Sets (Conjuntos)

Los sets son colecciones de elementos únicos (no permiten duplicados).

```python
# Ejemplo de conjunto (los duplicados se eliminan automáticamente)
conjunto = {1, 2, 3, 3, 2, 1, 4, 5, 5}

print(conjunto)  # Verás que queda {1, 2, 3, 4, 5}

# Agregar elementos
conjunto.add(6)                 # Agrega un elemento
print(conjunto)

# Agregar múltiples elementos (solo con update)
conjunto.update([7, 8, 9])      # Agrega varios elementos
print(conjunto)

# Eliminar elementos
conjunto.remove(2)              # Elimina un elemento específico (error si no existe)
print(conjunto)

conjunto.discard(3)             # Elimina sin error si no existe
print(conjunto)

eliminado = conjunto.pop()      # Elimina un elemento aleatorio
print("Elemento eliminado:", eliminado)
print(conjunto)

# Vaciar conjunto
copia_conjunto = conjunto.copy()  # Copia un set
conjunto.clear()                 # Limpia todo
print(conjunto)


# -----------------------------
# Operaciones entre conjuntos
# -----------------------------
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a.union(b))          # Unión
print(a | b)               # Operador de unión

print(a.intersection(b))   # Intersección
print(a & b)

print(a.difference(b))     # Diferencia
print(a - b)

print(a.symmetric_difference(b)) # Diferencia simétrica
print(a ^ b)

# Comprobaciones
print(a.issubset(b))       # ¿a es subconjunto de b?
print(a.issuperset(b))     # ¿a es superconjunto de b?
print(a.isdisjoint(b))     # ¿No tienen elementos en común?
```

### 📝 Problema 1: Procesamiento de Nombres

Imagina que recibes una cadena de texto con nombres de usuarios separados por comas, pero vienen con espacios extra y en diferentes formatos (mayúsculas/minúsculas). 

**Objetivo:** Limpiar los nombres, poner la primera letra en mayúscula y guardarlos en una lista ordenada.

```python
datos_sucios = "  jorge , ALBERTO,  maRia , lucia  "

# 1. Separar por comas (split)
nombres_lista = datos_sucios.split(",")

# 2. Limpiar espacios y capitalizar cada nombre
nombres_limpios = []
for n in nombres_lista:
    nombres_limpios.append(n.strip().capitalize())

# 3. Ordenar la lista final
nombres_limpios.sort()

print(f"Lista final ordenada: {nombres_limpios}")
```

### 📝 Problema 2: Análisis de Inventario Simple

Tienes una lista de productos vendidos en el día. Algunos se repiten. 

**Objetivo:** 
1. Obtener la lista de productos únicos vendidos.
2. Calcular cuántos productos totales se vendieron (incluyendo duplicados).
3. Crear un resumen usando un diccionario para el primer y último producto de la lista original.

```python
ventas = ["laptop", "mouse", "laptop", "teclado", "mouse", "monitor"]

# 1. Productos únicos (usando set)
productos_unicos = set(ventas)
print(f"Productos únicos: {productos_unicos}")

# 2. Cantidad total de ventas
total_ventas = len(ventas)
print(f"Total ventas realizadas: {total_ventas}")

# 3. Resumen en diccionario
resumen = {
    "primera_venta": ventas[0],
    "ultima_venta": ventas[-1],
    "variedad_productos": len(productos_unicos)
}
print(f"Resumen del día: {resumen}")
```
