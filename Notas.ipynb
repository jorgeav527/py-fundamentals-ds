```python
df = pd.read_csv("202309-7E PYTHON FUNDAMENTALS FOR DATA SCIENCE grades - 11_19_2023, 06_38 PM.csv")
```

```python
df = df.drop(11)
df = df.reset_index(drop=True)
```

```python
column_names_list = list(df.columns)
column_names_list
```

```python
column_name_mapping = {
    'Points': 'final_score', 
    'Points.1': 'score_7', 
    'Points.2': 'score_6',
    'Points.3': 'score_5', 
    'Points.4': 'score_4', 
    'Points.5': 'score_3',
    'Points.6': 'score_2',
    'Points.7': 'score_1',
    'Tarea - 02': 'Tarea-02',
    'Examen Final': 'Evaluacion Final (EF)',
}
```

```python
df = df.rename(columns=column_name_mapping)
```

```python
columns_to_keep = [
    'First Name', 'Last Name',
    'Evaluacion Final (EF)', 'final_score',
    'Tarea-07', 'score_7', 
    'Tarea-06', 'score_6', 
    'Tarea-05', 'score_5', 
    'Tarea-04', 'score_4', 
    'Tarea-03', 'score_3', 
    'Tarea-02', 'score_2', 
    'Tarea-01', 'score_1']
```

```python
notas_df = df[columns_to_keep]
notas_df
```

```python
notas_df.loc[9, 'Tarea-04'] = 3.5
# notas_df.loc[5, 'Tarea-04'] = 5
```

```python
# Calculate the average for the first 5 tasks
average_first_5 = notas_df[["Tarea-07", "Tarea-06", "Tarea-05", "Tarea-04", "Tarea-03"]].mean(axis=1)

# Convert the average to a grade base 20
grade_first_5 = average_first_5 * 20 / 5

# Add a new column named `grade_first_5` to the DataFrame using .loc
notas_df.loc[:, "Evaluacion Tareas (ET)"] = grade_first_5
```

```python
notas_df
```

```python
# Calculate the average for the first 2 tasks
average_first_2 = notas_df[["Tarea-02", "Tarea-01"]].mean(axis=1)

# Convert the average to a grade base 20
grade_first_2 = average_first_2 * 20 / 2

# Add a new column named `PP` to the DataFrame using .loc to avoid the warning
notas_df.loc[:, "Asistencia Practicas (PP)"] = grade_first_2
```

```python
notas_df
```

```python
notas_df.loc[:, 'Promedio Final (PF)'] = ((notas_df['Asistencia Practicas (PP)'] * 0.20) +
                                           (notas_df['Evaluacion Tareas (ET)'] * 0.30) +
                                           (notas_df['Evaluacion Final (EF)'] * 0.50)).round(2)
```
