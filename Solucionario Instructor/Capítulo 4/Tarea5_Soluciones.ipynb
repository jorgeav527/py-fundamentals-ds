```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
#1 - simular lanzamiento de una moneda con random (50 lanzamientos) y plotear resultados
# set the seed to get same random number
np.random.seed(42)
coin = [np.random.randint(0, 2) for _ in range(51)]
print(coin)
plt.plot(coin)
```

```python
#2 - simular lanzamiento con moneda con truco, cara sale el 70% de veces (50 lanzamientos) y plotear resultados
def flip(p):
    return '0' if random.random() < p else '1'
N = 51
flips = [flip(0.7) for i in range(N)]
print(flips)
print(float(flips.count('0'))/N)
plt.plot(flips)
```

```python
#3 - Crear un calendario 2020 en pandas, incluir como mínimo semana y día - puede usarse el inglés para los nombres
def create_cal(start='2020-01-01', end='2020-12-31'):
    df = pd.DataFrame({"Date": pd.date_range(start, end)})
    df["Day"] = df.Date.dt.weekday_name
    df["Week"] = df.Date.dt.weekofyear
    df["Quarter"] = df.Date.dt.quarter
    df["Year"] = df.Date.dt.year
    df["Year_half"] = (df.Quarter + 1) // 2
    return df
cal = create_cal()
cal
```
