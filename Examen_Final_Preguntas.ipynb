P9. Agrupar y contar los registros de <=50K y >50K. Graficar el resultado de manera apropiada

```python
conteo_salarios = df.groupby("salary").size()
```

```python
conteo_salarios
```

```python
import matplotlib.pyplot as plt

conteo_salarios.plot(kind="bar", color=["skyblue", "orange"])

plt.title("Conteo de registros por nivel de salario")
plt.xlabel("Salario")
plt.ylabel("Número de personas")
plt.xticks(rotation=0)
plt.show()
```
